# This file was used to generate box plots and scatter plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df =pd.read_csv("preprocessed.csv")
```

```python
df['Tenure']
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.boxplot(y=df['Tenure'])


plt.subplot(1, 3, 2)
sns.boxplot(y=df['Monthly Charges'])

plt.subplot(1, 3, 3)
sns.boxplot(y=df['Total Charges']/12)

plt.tight_layout()
plt.show()
```

```python
sampled_df = df.sample(n=2000, random_state=42)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=sampled_df, x='Tenure', y='Total Charges')
plt.title('Tenure vs Total Charges')
plt.show()
```
