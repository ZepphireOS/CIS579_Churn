```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../data/CustomerChurn.xlsx")
df.drop(columns=["LoyaltyID", "Customer ID"], inplace = True)
df
```

```python
bin_cols = [col for col in df if df[col].nunique()==2]
```

```python
bin_cols
```

```python

df[bin_cols] = df[bin_cols].replace({'Yes': True, 'No': False})
```

```python
trip_cols = ["Streaming Movies", "Streaming TV", "Tech Support", "Device Protection", "Online Backup", "Online Security", "Multiple Lines"]
df[trip_cols] = df[trip_cols].replace({"Yes": True, "No": False})
```

```python
df = df.applymap(
    lambda x: False if isinstance(x, str) and x.startswith("No ") else x
)
```

```python
df
```

Just a small revision to Bharath's onehot encoding, just that internet type can be FALSE and TRUE and if it is TRUE, the type of internet is listed.

```python
df['int_tf'] = df['Internet Service'].apply(lambda x: x != 'No')
df['int_type'] = df['Internet Service'].where(df['int_tf'], np.nan)
df = pd.get_dummies(df, columns=['int_type'])
```

```python
df
```

Same logic for the billing type/contract.

```python
df['yearly'] = df['Contract'] != 'Month-to-month'
df['Contract_Type'] = df['Contract'].where(df['yearly'], False)
```

```python
df = pd.get_dummies(df, columns=['Contract_Type'])
```

```python
df.drop(columns = "Contract_Type_False", inplace = True)
```

```python
df = pd.get_dummies(df, columns=["Payment Method"])
```

```python
df
```

```python
df.drop(columns=["Contract", "Internet Service"], inplace = True)
```
