```python
def dropman(df):
    if "LoyaltyID" in df.columns and "Customer ID" in df.columns:
        df.drop(columns=["LoyaltyID", "Customer ID"], inplace=True)
    else:
        df.drop(columns="CustomerID", , inplace=True)
    return df
```

```python
telco = dropman(telco)
cus = dropman(cus)
```
